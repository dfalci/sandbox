```
import nltk
nltk.download('mac_morpho')
nltk.download('floresta')

from nltk.corpus import floresta
```

```
t_corpus = floresta.tagged_sents()

novo_corpus = []

def tratar_tag(t):
    a = t[t.index("+")+1:] if "+" in t else t
    return a[:a.index('-')] if '-' in a else a

for t_sent in t_corpus:
    n_sent = []
    for t_word in t_sent:
        nova_tag = tratar_tag(t_word[1])
        t_word = (t_word[0].lower(), nova_tag)
        n_sent.append(t_word)
    novo_corpus.append(n_sent)

# print len(novo_corpus)
```

```
for i in xrange(0, 5):
    print novo_corpus[i]
```

```


count_tags = {}
count_trigrama = {}
count_bigrama = {}
count_w_tag = {}

teste = 0

def get_sequencia_hidden(seq, count):
    temp = ()
    for i in xrange(0, count):
        temp = temp + (seq[i][1], )
    return temp
    


for sent in novo_corpus:
    i = 1
    for w in sent:
        count_tags[w[1]] = count_tags.get(w[1], 0) +1
        count_w_tag[(w[0], w[1])] = count_w_tag.get((w[0], w[1]), 0) +1
        if i >= 3:
            trigrama = sent[i-3 : i]
            count_trigrama[get_sequencia_hidden(trigrama, 3)] = count_trigrama.get(get_sequencia_hidden(trigrama, 3), 0) + 1
            count_bigrama[get_sequencia_hidden(trigrama, 2)] = count_bigrama.get(get_sequencia_hidden(trigrama, 2), 0) + 1
        i+=1
    teste+=1
    if teste > 100000:
        break

print count_tags
#print count_w_tag
print count_trigrama
print count_bigrama[('v', 'n')]
```

```
def calcular_transicao(padrao):
    return count_trigrama.get(padrao, 1) / float(count_bigrama.get((padrao[0], padrao[1]), 1))

transicoes = (u'art', u'adj', u'n')
print calcular_transicao(transicoes)

print count_trigrama.get(transicoes, 1)
print count_bigrama.get((transicoes[0], transicoes[1]), 1)
    
        
```

```
def calcular_emissao(w, tag):
    return count_w_tag.get((w, tag), 1) / float(count_tags.get(tag, 1))

print calcular_emissao(u'ir', u'v')


print count_w_tag.get((u'dois', u'num'), 1)
print count_tags[u'num']
```

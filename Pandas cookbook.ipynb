```python
df = pd.read_csv("boston_train.csv",sep=',')
```

```python
df[['crim', 'rm']].head() # captura um subset que contem apenas as colunas desejadas
```

```python
df.loc[:2] # buscando de acordo com o indice (primeira coluna sem nome)
```

```python
df.iloc[:2] # buscando de acordo com a quantidade de linhas... note a diferença... mais utilizado
```

```python
df.iloc[-5:]# ultimas 5 linhas
```

## Filtrando elementos

```python
df[df['crim'] > 25] # itens com taxa de criminalidade superior a 25
```

```python
df[(df['crim'] > 35) & (df['age'] < 100)] # que tem alta criminalidade e menos de 100 anos
```

```python
df[df['lstat'] > df['medv'] ].head() # registros cujo valor seja maior que outro dentro da mesma linha
```

## Removendo elementos de um dataset

```python
# Remove as colunas indicadas no array. Neste caso apenas a coluna zn. 
# O parametro axis == 1 indica a coluna ao passo que axis == 0 indica linha
# note ainda que df não é alterado definitivamente. 
df.drop(['zn'], axis=1).head() 
```

## Aplicando função a colunas e linhas

```python
# aplica uma função aos elementos das colunas abaixo... col vem com o array completo.... 
# o parametro axis neste caso é 0 porque os elementos selecionados são dispostos em arrays empilhados

def col_func(col):
    return np.mean(col)

df[['crim',"chas"]].apply(lambda col: col_func(col),axis=0)
```

```python
# aplica uma funcao aos elementos de uma linha qualquer.... row vem com todos os registros da linha

def row_func(row):
    return row['crim'] * 2


df.apply(lambda row: row_func(row), axis=1).head()
```

## Adicionando colunas baseadas em outras colunas

```python
df['age_2'] = (df['age'] * 2)
df.head()
```

```python
df['nome'] = df.apply(lambda row: 'idade {}'.format(row['age_2']), axis=1)
df.head()
```

```python
# outra opção para transformar os valores das celulas de uma coluna é utilizar o map
df['nome2'] = df['nome'].map(lambda nome: nome.upper())
df.head()
```

## Randomizar as linhas do dataframe

```python
# note que os indices originais são preservados
df.reindex(np.random.permutation(df.index)).head() 
```

## Iterar nas linhas do dataframe diretamente

```python
for indice, row in df.iterrows():
    print 'Linha com indice {} tem o nome {} e a taxa de criminalidade {}'.format(indice,row["nome"], row['crim'])
    if indice >= 5:
        break
```

## Criando dataframes a partir de dicionarios

```python
novo = pd.DataFrame(columns=['id','nome', 'valor'])
dicionario = [
    {'id':1,'nome': 'registro 1', 'valor':10.},
    {'id':2,'nome':'registro 2', 'valor':11.},
    {'id':3,'nome': 'registro 3', 'valor':12.},
]
novo = novo.from_records(dicionario)
novo.head()
```
